```python
import pandas as pd
import numpy as np

df = pd.read_csv("D:/Microsoft Data Engineer/Graduation Project/Database/cleaned_online_retail.csv")
```

```python
df.head()
```

```python
# Create customers.csv
customers = df[['CustomerID', 'Country']].drop_duplicates(subset=['CustomerID'])
customers.to_csv('D:/Microsoft Data Engineer/Graduation Project/Database/customers.csv', index=False)

# Create products.csv
products = df[['StockCode', 'Description', 'UnitPrice']].drop_duplicates(subset=['StockCode'])
products.to_csv('D:/Microsoft Data Engineer/Graduation Project/Database/products.csv', index=False)

# Create invoices.csv
invoices = df[['InvoiceNo', 'CustomerID', 'InvoiceDate']].drop_duplicates(subset=['InvoiceNo'])
invoices.to_csv('D:/Microsoft Data Engineer/Graduation Project/Database/invoices.csv', index=False)

# Create invoice_items.csv (ensure all InvoiceNo values exist in invoices.csv)
invoice_items = df[['InvoiceNo', 'StockCode', 'Quantity', 'TotalPrice']]
invoice_items = invoice_items[invoice_items['InvoiceNo'].isin(invoices['InvoiceNo'])]
invoice_items.to_csv('D:/Microsoft Data Engineer/Graduation Project/Database/invoice_items.csv', index=False)
```
